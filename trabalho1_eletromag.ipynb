```python
#importando as bibliotecas necess√°rias 
import numpy as np
import pandas as pd 
import math 
import matplotlib.pyplot as plt 
```

```python
#quest√£o 1: intensidade de campo magn√©tico em torno de um fio ret√≠lineo e infinito 
vet =[]
vetp = []
I = 10 #corrente
for p in range(1,11):
    h = I/(2*3.14*p) #aplica√ß√£o de Biot-Savart
    vetp.append(p)
    vet.append(h)
plt.figure(figsize=(10,5))
plt.plot(vetp,vet,lw=3)
plt.title("Intensidade de campo magn√©tico num fio retilineo infinito x Dist√¢ncia", size = 14)
plt.xlabel("Dist√¢ncia (m)",size=12)
plt.ylabel("Intensidade (A/m)",size=12)
plt.grid(color='gray', linestyle='-', linewidth=1)
```

```python
#quest√£o 2: intensidade do campo magn√©tico em um fio retil√≠neo finito obs: considerando que p esteja no eixo y. 
vet1 = []
vetp = []
for p in range(1,11):
    vetp.append(p)
    sen2 = 15/math.sqrt(15**2 + p**2)
    sen1 = -15/math.sqrt(15**2 + p**2)
    h = (I/(4*3.14*p))*(sen2-sen1)
    vet1.append(h)
plt.figure(figsize=(10,5))
plt.plot(vetp,vet1,lw=3)
plt.title("Intensidade de campo magn√©tico num fio retilineo finito x Dist√¢ncia", size = 14)
plt.xlabel("Dist√¢ncia (m)",size=12)
plt.ylabel("Intensidade (A/m)",size=12)
plt.grid(color='gray', linestyle='-', linewidth=1)
```
