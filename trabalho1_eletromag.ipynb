```python
#importando as bibliotecas necessárias 
import numpy as np
import pandas as pd 
import math 
import matplotlib.pyplot as plt 
```

```python
#questão 1: intensidade de campo magnético em torno de um fio retílineo e infinito 
vet =[]
vetp = []
I = 10 #corrente
for p in range(1,11):
    h = I/(2*3.14*p) #aplicação de Biot-Savart
    vetp.append(p)
    vet.append(h)
plt.figure(figsize=(10,5))
plt.plot(vetp,vet,lw=3)
plt.title("Intensidade de campo magnético num fio retilineo infinito x Distância", size = 14)
plt.xlabel("Distância (m)",size=12)
plt.ylabel("Intensidade (A/m)",size=12)
plt.grid(color='gray', linestyle='-', linewidth=1)
```

```python
#questão 2: intensidade do campo magnético em um fio retilíneo finito obs: considerando que p esteja no eixo y. 
vet1 = []
vetp = []
for p in range(1,11):
    vetp.append(p)
    sen2 = 15/math.sqrt(15**2 + p**2)
    sen1 = -15/math.sqrt(15**2 + p**2)
    h = (I/(4*3.14*p))*(sen2-sen1)
    vet1.append(h)
plt.figure(figsize=(10,5))
plt.plot(vetp,vet1,lw=3)
plt.title("Intensidade de campo magnético num fio retilineo finito x Distância", size = 14)
plt.xlabel("Distância (m)",size=12)
plt.ylabel("Intensidade (A/m)",size=12)
plt.grid(color='gray', linestyle='-', linewidth=1)
```
